```python
# Import Dependencies

import pandas as pd
```

```python
# Import schools and student data file and evaluate

schools = ".git/schools_complete.csv"
schools_df = pd.read_csv(schools)

# Review schools_df data (disabled)
# schools_df

students = ".git/students_complete.csv"
students_df = pd.read_csv(students)

# Review students_df data (disabled)
# students_df
```

```python
# Merge student and schools datafiles to create single data file

merge_df = pd.merge(schools_df, students_df, on="school_name", how="outer")

# Review merged file; evaluate missing values or data, evaluate datatype (disabled)
# merge_df.head()
# merge_df.count()
# merge_df.dtypes
```

```python
## District Summary:  Create a high level snapshot in table form of the district's key metrics including:

# Step 1:  Calculate Total Schools, Total Students, Total Budget, Average Math Scores, Average Reading Score

TotSchools = schools_df["School ID"].count()
TotStudents = merge_df["Student ID"].count()
TotBudget = schools_df["budget"].sum() 
AvgMath = merge_df["math_score"].mean()
AvgRead = merge_df["reading_score"].mean()

# Step 2:  Calculate % Passing Math, % Passing Reading, and % Overall Passing (defined as passing Math and Reading)

# Step 2A:  Created dataframes filtered by passing scores (TotPass_df filters ReadPass_df for math score >= 70)

MathPass_df = merge_df[merge_df["math_score"]>=70]
ReadPass_df = merge_df[merge_df["reading_score"]>=70]
TotPass_df = ReadPass_df[ReadPass_df["math_score"]>=70]

# Step 2B:  Count total passing students and divide by Total Students

MathPass = MathPass_df["math_score"].count()
ReadPass = ReadPass_df["reading_score"].count()
TotPass = TotPass_df["math_score"].count()
MPerPass = round(MathPass / TotStudents, 4)*100
RPerPass = round(ReadPass / TotStudents, 4)*100
TPerPass = round(TotPass / TotStudents, 4)*100

# Step 3:  Define output and headers and incoporate into DataFrame

Output = [TotSchools, TotStudents, TotBudget, AvgMath, AvgRead, MPerPass, RPerPass, TPerPass]
Headers = ["Total Schools", "Total Students", "Total Budget", "Average Math Score", "Average Reading Score", 
           "% Passing Math", "% Passing Reading", "% Overall Passing"]

District_df = pd.DataFrame(Output, Headers).transpose()
District_df
```

```python
## School Summary:  Create an overview table that summarizes key metrics about each school

# Step 1: Create dataframe indexed on school names with school types, use "schools_df" to start

SchoolType_df = schools_df.loc[:,["school_name", "type"]]
SchoolType_df = SchoolType_df.set_index("school_name")
SchoolType_df = SchoolType_df.rename(columns={"type": "School Type"}) 

# Review output (disabled)
# SchoolType_df.head()
```

```python
# Step 2:  Calculate Total Students and Total School Budget

SchoolStu = merge_df.groupby("school_name")["size"].mean()
SchoolBud = merge_df.groupby("school_name")["budget"].mean()

SchoolStu_df = pd.DataFrame({"Total Students": SchoolStu})
SchoolBud_df = pd.DataFrame({"Total School Budget":  SchoolBud})

# Review output (disabled)
# SchoolBud_df
```

```python
# Step 3:  Calcualte Per Student Budget, Average Math, and Average Reading Scorres

AvgMath1 = round(merge_df.groupby("school_name")["math_score"].mean(), 2) 
AvgRead1 = round(merge_df.groupby("school_name")["reading_score"].mean(), 2)
BudPerStu = merge_df.groupby("school_name")["budget"].mean() / merge_df.groupby("school_name")["size"].mean()

BudPerStu_df = pd.DataFrame({"Per Student Budget":  BudPerStu})
AvgRead1_df = pd.DataFrame({"Average Reading Score": AvgRead1})
AvgMath1_df = pd.DataFrame({"Average Math Score": AvgMath1})

# Review output (disabled)
# BudPerStu_df
```

```python
# Step 4:  Create filtered dataframes (Trial1_df, etc.), to hold count values for passing students (greater than 70%)
    # Divide filtered dataframes by "merge_df", grouped by school, using count function to determine passing %'s 
    # Note:  Trial3 out refers to TotPass_df, which is ReadPass filtered by pathing mass scores for total passing data

Trial1_df = MathPass_df.groupby("school_name").count()
Trial2_df = ReadPass_df.groupby("school_name").count()
Trial3_df = TotPass_df.groupby("school_name").count()

MathPass1 = round(Trial1_df["math_score"] / merge_df.groupby("school_name")["math_score"].count(), 2) * 100
ReadPass1 = round(Trial2_df["reading_score"] / merge_df.groupby("school_name")["reading_score"].count(), 2) * 100
TotPass1 = round(Trial3_df["reading_score"] / merge_df.groupby("school_name")["reading_score"].count(), 2) * 100

MathPass1_df = pd.DataFrame({"% Passing Math": MathPass1}) 
ReadPass1_df = pd.DataFrame({"% Passing Reading":  ReadPass1})
TotPass1_df = pd.DataFrame({"% Overall Passing":  TotPass1})

# Review output (disabled)
# School_ReadPass_df.head()
```

```python
# Step 5:  Re-assemble using the concatenate function

columns = [SchoolType_df, SchoolStu_df, SchoolBud_df, BudPerStu_df, AvgMath1_df, AvgRead1_df, MathPass1_df, 
           ReadPass1_df, TotPass1_df]
SchoolSum1_df = pd.concat(columns, axis=1)
SchoolSum1_df
```

```python
## Top Performing Schools, use sort function descending

SchoolSum1_df.sort_values(by="% Overall Passing", ascending=False).head()
```

```python
## Bottom Performing Schools, use sort function ascending

SchoolSum1_df.sort_values(by="% Overall Passing", ascending=True).head()
```

```python
## Math Scored by Grade

# Step 1:  Seperate merge_df by grade
Nine_in_df = merge_df[merge_df["grade"] == "9th"]
Ten_in_df = merge_df[merge_df["grade"] == "10th"]
Eleven_in_df = merge_df[merge_df["grade"] == "11th"]
Twelve_in_df = merge_df[merge_df["grade"] == "12th"]

# Step 2:  Group separated files by shool and calculate mean values
Nine = round(Nine_in_df.groupby("school_name")["math_score"].mean(), 2)
Ten = round(Ten_in_df.groupby("school_name")["math_score"].mean(), 2)
Eleven = round(Eleven_in_df.groupby("school_name")["math_score"].mean(), 2)
Twelve = round(Twelve_in_df.groupby("school_name")["math_score"].mean(), 2)

# Step 3:  Compile DataFrames
Nine_out_df = pd.DataFrame({"9th Grade":  Nine})
Ten_out_df = pd.DataFrame({"10th Grade":  Ten})
Eleven_out_df = pd.DataFrame({"11th Grade":  Eleven})
Twelve_out_df = pd.DataFrame({"12th Grade":  Twelve})

#Step 4:  Combine and output
MathSum_df = pd.concat([Nine_out_df, Ten_out_df, Eleven_out_df, Twelve_out_df], axis=1)
MathSum_df
```

```python
## Reading Scores by Grade


# Step 1:  Seperate merge_df by grade
Nine1_in_df = merge_df[merge_df["grade"] == "9th"]
Ten1_in_df = merge_df[merge_df["grade"] == "10th"]
Eleven1_in_df = merge_df[merge_df["grade"] == "11th"]
Twelve1_in_df = merge_df[merge_df["grade"] == "12th"]

# Step 2:  Group separated files by shool and calculate mean values
Nine1 = round(Nine1_in_df.groupby("school_name")["reading_score"].mean(), 2)
Ten1 = round(Ten1_in_df.groupby("school_name")["reading_score"].mean(), 2)
Eleven1 = round(Eleven1_in_df.groupby("school_name")["reading_score"].mean(), 2)
Twelve1 = round(Twelve1_in_df.groupby("school_name")["reading_score"].mean(), 2)

# Step 3:  Compile DataFrames
Nine1_out_df = pd.DataFrame({"9th Grade":  Nine1})
Ten1_out_df = pd.DataFrame({"10th Grade":  Ten1})
Eleven1_out_df = pd.DataFrame({"11th Grade":  Eleven1})
Twelve1_out_df = pd.DataFrame({"12th Grade":  Twelve1})

#Step 4:  Combine and output
MathSum_df = pd.concat([Nine1_out_df, Ten1_out_df, Eleven1_out_df, Twelve1_out_df], axis=1)
MathSum_df
```

```python
## Scores by School Spending

# Create bins for Student Spending; add to schoolsum_df
bins = [0, 583, 628, 643, 676]
group_name = ["< 584", "$585-629", "$630-644", "645-675"]          
SchoolSum1_df["Spending Ranges (per Student)"] = pd.cut(SchoolSum1_df["Per Student Budget"], bins, labels=group_name, 
                                                       include_lowest=True)

# Group values by Spending bins; calculate mean values
spendsum_df = round(SchoolSum1_df.groupby(["Spending Ranges (per Student)"]).mean(), 2)
spendsum_df = spendsum_df[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading",
                          "% Overall Passing"]]
spendsum_df
```

```python
## Scores by School Size

# Create bins for School size, add to schoolsum_df
bins1 = [0, 1000, 1999, 5001]
group_name1 = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
SchoolSum1_df["School Size"] = pd.cut(SchoolSum1_df["Total Students"], bins1, labels=group_name1, include_lowest=True)

# Group values by Size bins; calucate mean values
sizesum_df = round(SchoolSum1_df.groupby(["School Size"]).mean(), 2)
sizesum_df = sizesum_df[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading",
                          "% Overall Passing"]]
sizesum_df.head()
```

```python
## Scores by School Type

#  Group schools by type; calculate mean values
typesum_df = round(SchoolSum1_df.groupby(["School Type"]).mean(), 2)
typesum_df = typesum_df[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading",
                          "% Overall Passing"]]
typesum_df 
```

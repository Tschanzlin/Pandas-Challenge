```python
# Import Dependencies

import pandas as pd
```

```python
# Import schools and student data file and evaluate

schools = ".git/schools_complete.csv"
schools_df = pd.read_csv(schools)

# Review schools_df data (disabled)
# schools_df

students = ".git/students_complete.csv"
students_df = pd.read_csv(students)

# Review students_df data (disabled)
# students_df
```

```python
# Merge student and schools datafiles to create single data file

merge_df = pd.merge(schools_df, students_df, on="school_name", how="outer")

# Review merged file; evaluate missing values or data, evaluate datatype (disabled)
# merge_df.head()
# merge_df.count()
# merge_df.dtypes
```

```python
## District Summary:  Create a high level snapshot in table form of the district's key metrics including:

# Step 1:  Calculate Total Schools, Total Students, Total Budget, Average Math Scores, Average Reading Score

TotSchools = schools_df["School ID"].count()
TotStudents = merge_df["Student ID"].count()
TotBudget = schools_df["budget"].sum() 
AvgMath = merge_df["math_score"].mean()
AvgRead = merge_df["reading_score"].mean()

# Step 2:  Calculate % Passing Math, % Passing Reading, and % Overall Passing (defined as passing Math and Reading)

# Step 2A:  Created dataframes filtered by passing scores (TotPass_df filters ReadPass_df for math score >= 70)

MathPass_df = merge_df[merge_df["math_score"]>=70]
ReadPass_df = merge_df[merge_df["reading_score"]>=70]
TotPass_df = ReadPass_df[ReadPass_df["math_score"]>=70]

# Step 2B:  Count total passing students and divide by Total Students

MathPass = MathPass_df["math_score"].count()
ReadPass = ReadPass_df["reading_score"].count()
TotPass = TotPass_df["math_score"].count()
MPerPass = round(MathPass / TotStudents, 4)*100
RPerPass = round(ReadPass / TotStudents, 4)*100
TPerPass = round(TotPass / TotStudents, 4)*100

# Step 3:  Define output and headers and incoporate into DataFrame

Output = [TotSchools, TotStudents, TotBudget, AvgMath, AvgRead, MPerPass, RPerPass, TPerPass]
Headers = ["Total Schools", "Total Students", "Total Budget", "Average Math Score", "Average Reading Score", 
           "% Passing Math", "% Passing Reading", "% Overall Passing"]

District_df = pd.DataFrame(Output, Headers).transpose()
District_df
```

```python
## School Summary:  Create an overview table that summarizes key metrics about each school

# Step 1: Create dataframe indexed on school names with school types, use "schools_df" to start

schooltype_df= schools_df.loc[:,["school_name", "type"]]
schooltype_df = schooltype_df.set_index("school_name")
schooltype_df = schooltype_df.rename(columns={"type": "School Type"}) 

# schooltype_df
```

```python
# Step 2:  Create dataframes grouped by (and indexed by) "school_name" to calculate values by school 

# Step 2A:  Create dataframe "group_df" to hold counts for all students 
    # Will be used for "Total Students" and to calculate % passing scores

group_df = merge_df.groupby("school_name").count()

# Review output (disabled)
# group_df.head()
```

```python
# Step 2B:  Create filtered dataframes, to hold count values for passing students (greater than 70%)
    # Counting passing students in MathPass, ReadPass, and TotPass dataframes (passing grades), grouped by school
    # Note:  Trial3 out refers to TotPass_df, which is ReadPass filtered by MathPass for Total Passing data

Trial1_df = MathPass_df.groupby("school_name").count()
Trial2_df = ReadPass_df.groupby("school_name").count()
Trial3_df = TotPass_df.groupby("school_name").count()
```

```python
# Step 2C:  Calcluate % passing math, reading and overall scores by dividing filtered values by total values

Trial1out = Trial1_df.loc[:, "math_score"] / group_df.loc[:, "math_score"]
Trial2out = Trial2_df.loc[:, "reading_score"] / group_df.loc[:, "reading_score"]
Trial3out = Trial3_df.loc[:, "reading_score"] / group_df.loc[:, "reading_score"]

# Review calculated values against expected output (disabled)
# Trial1out
```

```python
# Step 2D:  Convert above files to dataframes, store values under proper heading names and indexed by "school_type"

Trial1out_df = pd.DataFrame({"% Passing Math": Trial1out})
Trial2out_df = pd.DataFrame({"% Passing Reading": Trial2out})
Trial3out_df = pd.DataFrame({"% Overall Passing": Trial3out})

# Review output (disabled)
# Trial1out_df.head()
```

```python
# Step 3:  Calculate Total Students, Total School Budget, Average Math Score, Average Reading Score
    # Total Students and Total Budget are aveage values of groupby "school_type" dataframe
    # Create group1_df to hold mean values, indexed by school; select relevant data

group1_df = merge_df.groupby("school_name").mean()
maingroup_df = group1_df
maingroup_df = maingroup_df[["size", "budget", "reading_score", "math_score"]]
maingroup_df = maingroup_df.rename(columns={"size": "Total Students", "budget":"Total School Budget", 
                                            "reading_score":"Average Reading Score", "math_score": "Average Math Score"})
# Review output (disabled)
# maingroup_df.head()
```

```python
# Step 4:  Calcualte Per Student Budget

budperstu = maingroup_df.loc[:, "Total School Budget"] / maingroup_df.loc[:, "Total Students"]
# budperstu
budperstu_df = pd.DataFrame({"Per Student Budget":budperstu})

# Review output (disabled)
# budperstu_df.head()
```

```python
# Step 5:  Re-assemple using the concatenation function

schoolsum_df = pd.concat([schooltype_df, maingroup_df, budperstu_df, Trial1out_df, Trial2out_df, Trial3out_df], axis=1)

# Reorganize and format columns
schoolsum_df = schoolsum_df[["School Type", "Total Students", "Total School Budget", "Per Student Budget", 
                            "Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading",
                            "% Overall Passing"]]

schoolsum_df["Total School Budget"] = schoolsum_df["Total School Budget"].map("${:,.2f}".format)
#schoolsum_df["Per Student Budget"] = schoolsum_df["Per Student Budget"].map("${:.2f}".format)
#schoolsum_df["Average Math Score"] = schoolsum_df["Average Math Score"].map("{:.2f}%".format)
#schoolsum_df["Average Reading Score"] = schoolsum_df["Average Reading Score"].map("{:.2f}%".format)
schoolsum_df

# COMMENTS:  Reformatting changed Per Student Budge from float64 to object, which caused scores by school spending
    # to fail and other "object" series to fall-out. Disabled those formatting efforts.  Also unable to alphabetically 
    # sort the index
```

```python
## Scores by School Spending

# Create bins for Student Spending; add to schoolsum_df
bins = [0, 583, 628, 643, 676]
group_name = ["< 584", "$585-629", "$630-644", "645-675"]          
schoolsum_df["Spending Ranges (per Student)"] = pd.cut(schoolsum_df["Per Student Budget"], bins, labels=group_name, 
                                                       include_lowest=True)

# Group values by Spending bins; calculate mean values
spendsum_df = schoolsum_df.groupby(["Spending Ranges (per Student)"]).mean()
spendsum_df = spendsum_df[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading",
                          "% Overall Passing"]]
spendsum_df
```
